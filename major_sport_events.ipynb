```python
import pandas as pd

# Si tu es bien dans le repo (Israel-Palestine) et que le fichier est Ã  la racine :
df_wc = pd.read_csv("FifaWorldCup_raw.csv")

print("shape:", df_wc.shape)
print("\ncolumns:", list(df_wc.columns))




```

```python
# 1) Valeurs uniques de tournament_name
print(df_wc["tournament_name"].unique())
print("\n")
```

```python
#keep only FIFA men's World Cup
df_wc_men = df_wc[df_wc["tournament_name"].str.contains("Men's World Cup", regex=False)].copy()
df_wc_men["tournament_name"].unique()

```

```python
# 2) Valeurs uniques de stage_name
print(df_wc["stage_name"].unique())
print("\n")
```

```python
#harmoniser le nom des stages

df_wc_men["stage_name_clean"] = (
    df_wc_men["stage_name"]
    .str.lower()
    .str.strip()
)

#garder uniquement demi-finales et finales
keep_stages = [
    "semi-finals",
    "semi-final",
    "final"
]

df_wc_major = df_wc_men[
    df_wc_men["stage_name_clean"].isin(keep_stages)
].copy()

cols_keep = [
    "match_date",
    "tournament_name",
    "stage_name",
    "match_name",
]

df_wc_major = df_wc_major[cols_keep].copy()

df_wc_major.head()
```

```python
df_wc_major["match_date"].unique()

#filter the date

df_wc_major["match_date"] = pd.to_datetime(df_wc_major["match_date"])

df_wc_2000_2011 = df_wc_major[
    (df_wc_major["match_date"] >= "2000-01-01") &
    (df_wc_major["match_date"] <= "2011-12-31")
].copy()

if "match_date" in df_wc_2000_2011.columns:
    df_wc_2000_2011 = df_wc_2000_2011.rename(columns={"match_date": "date"})
df_wc_2000_2011["date"] = pd.to_datetime(df_wc_2000_2011["date"])



# --- save dataset 2000-2011 ---
out_path = "fifa_major_events_2000_2011.csv"
df_wc_2000_2011.to_csv(out_path, index=False)

df_wc_2000_2011.head()

```

```python
#filter the date

df_wc_major["match_date"] = pd.to_datetime(df_wc_major["match_date"])

df_wc_2000_2019 = df_wc_major[
    (df_wc_major["match_date"] >= "2000-01-01") &
    (df_wc_major["match_date"] <= "2019-12-31")
].copy()

if "match_date" in df_wc_2000_2019.columns:
    df_wc_2000_2019 = df_wc_2000_2019.rename(columns={"match_date": "date"})
df_wc_2000_2019["date"] = pd.to_datetime(df_wc_2000_2019["date"])

# --- save dataset 2000-2011 ---
out_path = "fifa_major_events_2000_2019.csv"
df_wc_2000_2019.to_csv(out_path, index=False)
df_wc_2000_2019.head()
```
